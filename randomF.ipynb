```python
# 📦 导入依赖
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

# ✅ 准备 pats_df 的 ID
pats_df.columns = pats_df.columns.str.strip()
pats_df['ID'] = pats_df['Id Patient V2018'].str.strip().str.lower()

# ✅ 合并
merged_full = merged_df.merge(pats_df, left_on="Patient_ID", right_on="ID", how="left")

# ✅ 构造 Age 特征
merged_full['Date of birth'] = pd.to_datetime(merged_full['Date of birth'], errors='coerce')
merged_full['Visit Date'] = pd.to_datetime(merged_full['Visit Date'], errors='coerce')
merged_full['Age'] = ((merged_full['Visit Date'] - merged_full['Date of birth']).dt.days / 365.25).round(1)

# ✅ 人工构造合并后的 Race 信息（如原为 one-hot 可简化为 dominant label）
def combine_race(row):
    if row['Race white'] == 'Yes':
        return 'White'
    elif row['Race black'] == 'Yes':
        return 'Black'
    elif row['Race asian'] == 'Yes':
        return 'Asian'
    elif row['Any other white'] == 'Yes':
        return 'Other'
    elif row['Hispanic'] == 'Yes':
        return 'Hispanic'
    else:
        return 'Unknown'

merged_full['Race'] = merged_full.apply(combine_race, axis=1)

# ✅ 选择特征
selected_features = [
    'Forced Vital Capacity (FVC - % predicted)',
    'BNP (pg/ml)', 'NTproBNP (pg/ml)',
    'Modified Rodnan Skin Score, only imported value',
    '6 Minute walk test (distance in m)',
    'Dyspnea (NYHA-stage)',
    'TAPSE: tricuspid annular plane systolic excursion in cm',
    'Extent of skin involvement',
    'HRCT: Lung fibrosis',
    'Sex', 'Height', 'Race',
    'Subsets of SSc according to LeRoy (1988)', 'Age'
]

df_model = merged_full[selected_features + ['Scleroderma present']].copy()

# ✅ 编码目标变量
df_model['target'] = df_model['Scleroderma present'].map({'No': 0, 'Yes': 1})

# ✅ 填补缺失值
num_cols = df_model.select_dtypes(include=['float64', 'int64']).columns.drop('target')
cat_cols = df_model.select_dtypes(include=['object']).columns.drop('Scleroderma present')

imputer = SimpleImputer(strategy='median')
df_model[num_cols] = imputer.fit_transform(df_model[num_cols])

df_model[cat_cols] = df_model[cat_cols].fillna('Unknown')

# ✅ 编码类别变量
le = LabelEncoder()
for col in cat_cols:
    df_model[col] = le.fit_transform(df_model[col])

# ✅ 拆分数据集
X = df_model.drop(columns=['Scleroderma present', 'target'])
y = df_model['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# ✅ SMOTE 重采样
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)

# ✅ 训练随机森林
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_res, y_res)

# ✅ 预测与评估
y_prob = rf.predict_proba(X_test)[:, 1]
y_pred = (y_prob >= 0.25).astype(int)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.title("Confusion Matrix (threshold=0.25)")
plt.show()

print("🎯 ROC-AUC:", roc_auc_score(y_test, y_prob))

# ✅ 特征重要性可视化
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=importances.index)
plt.title("Feature Importances")
plt.tight_layout()
plt.show()
```
